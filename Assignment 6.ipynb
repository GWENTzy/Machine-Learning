#### Discuss to find the correlation and covariance of the data (the variables used can be categorical vs categorical, categorical vs numeric, or numeric vs numeric variables).

#### Correlation

```python
from sklearn.preprocessing import LabelEncoder
```

```python
temp_dataset = dataset.copy()
encoder = LabelEncoder()
temp_dataset['name'] = encoder.fit_transform(temp_dataset['name'])
temp_dataset['host_name'] = encoder.fit_transform(temp_dataset['host_name'])
temp_dataset['neighbourhood_group'] = encoder.fit_transform(temp_dataset['neighbourhood_group'])
temp_dataset['neighbourhood'] = encoder.fit_transform(temp_dataset['neighbourhood'])
temp_dataset['room_type'] = encoder.fit_transform(temp_dataset['room_type'])
temp_dataset['last_review'] = encoder.fit_transform(temp_dataset['last_review'])
```

```python
temp_dataset.head()
```

```python
plt.figure(figsize=(14, 14))
sns.heatmap(temp_dataset.corr(), annot=True)

plt.tight_layout()
plt.show()
```

#### Covariance

```python
plt.figure(figsize=(16, 16))
sns.heatmap(temp_dataset.cov(), vmin=0, vmax=1, annot=True)

plt.tight_layout()
plt.show()
```
